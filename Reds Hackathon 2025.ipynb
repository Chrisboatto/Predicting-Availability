```python
# Define the list of events we're interested in
events_of_interest = [
    'strikeout', 'field_out', 'walk', 'force_out', 'sac_fly', 'single', 'hit_by_pitch', 
    'double', 'grounded_into_double_play', 'sac_bunt', 'home_run', 'fielders_choice', 
    'field_error', 'other_out', 'triple', 'strikeout_double_play', 'fielders_choice_out', 
    'double_play', 'sac_fly_double_play', 'catcher_interf', 'triple_play', 'sac_bunt_double_play'
]

# Assuming 'events' column contains the event type and 'batterID' and 'year' exist in your dataset
# Loop through the dataset (assuming the name of the DataFrame is `savant_data` or similar)
savant_data['pa'] = 0  # Initialize the column

# Loop through each year to calculate the number of unique batterID instances for each event
for year in savant_data['game_year'].unique():  # Iterate over all unique years
    # Filter data for the given year and events of interest
    year_data = savant_data[(savant_data['game_year'] == year) & 
                             (savant_data['events'].isin(events_of_interest))]
    
    # Group by batterID and count unique instances
    unique_batter_counts = year_data.groupby('batter').size()
    
    # Assign the count back to the original DataFrame for each batterID
    savant_data.loc[savant_data['game_year'] == year, 'pa'] = \
        savant_data.loc[savant_data['game_year'] == year, 'batter'].map(unique_batter_counts).fillna(0).astype(int)

# Display the DataFrame with the new column
print(savant_data[['game_year', 'batter', 'pa']].head())
```

## Merging the Data

This gives me the entire universe to use within the model.

```python
def merge_with_lahman(savant_data, lahman_people, merge_columns, suffix_map):
    """
    Merges savant_data with lahman_people based on a list of columns, 
    adding suffixes for each merge.

    Parameters:
        savant_data (DataFrame): The main dataset (e.g., savant_data).
        lahman_people (DataFrame): The secondary dataset (e.g., lahman_people).
        merge_columns (list): List of column names in savant_data to merge with player_mlb_id.
        suffix_map (dict): A mapping of column names to suffixes for merged columns.

    Returns:
        DataFrame: A single DataFrame with merged data.
    """
    df = savant_data.copy()  # Start with the main dataset
    
    for col in merge_columns:
        suffix = suffix_map.get(col, '')  # Get suffix for the current column
        df = pd.merge(
            df,
            lahman_people[['player_mlb_id', 'birthYear', 'birthMonth', 'birthDay', 'birthCountry', 
                           'weight', 'height', 'bats', 'throws', 'debut', 'birthDate']],
            left_on=col,
            right_on='player_mlb_id',
            how='left',
            suffixes=('', suffix)
        )
        df.drop(columns=['player_mlb_id'], inplace=True)  # Remove redundant column
    
    return df

# Define the columns to merge and their suffixes
merge_columns = ['pitcher', 'batter']
suffix_map = {'pitcher': '_pitcher', 'batter': '_batter'}

# Use the function to merge the datasets
df = merge_with_lahman(savant_data, lahman_people, merge_columns, suffix_map)

# Print the result
print("Merged DataFrame:")
print(df)
```

```python
def remove_columns_with_40_chars_except_batter(df):
    """
    Removes columns from the DataFrame where all values are strings of exactly 40 characters,
    except for the 'batter' column.
    """
    for column in df.columns:
        # Check if all values in the column are strings with exactly 40 characters
        if df[column].apply(lambda x: isinstance(x, str) and len(x) == 40).all():
            # Drop the column if the condition is met
            df.drop(column, axis=1, inplace=True)
    return df

# Apply the function to the pitcher and hitter datasets
df = remove_columns_with_40_chars_except_batter(df)
df.head(10)
```

```python
def calculate_age(df, game_year_column='game_year', birth_date_columns=['birthDate_pitcher', 'birthDate_batter']):
    """
    This function calculates the age of a player based on their birthDate and the game_year.
    Adds new columns 'age_pitcher' and 'age_hitter' to the DataFrame.
    
    Parameters:
        df (DataFrame): The DataFrame containing player data.
        game_year_column (str): The column name that contains the game year.
        birth_date_columns (list): List of birth date column names to calculate age for.
        
    Returns:
        df (DataFrame): The DataFrame with new 'age_pitcher' and 'age_hitter' columns.
    """
    for birth_date_column in birth_date_columns:
        # Ensure birthDate is in datetime format
        df[birth_date_column] = pd.to_datetime(df[birth_date_column], errors='coerce')
        
        # Extract the suffix from the column name (e.g., 'pitcher' or 'hitter')
        suffix = birth_date_column.split('_')[-1]
        
        # Calculate age based on birthDate and game_year (using first day of game year)
        df[f'age_{suffix}'] = df.apply(
            lambda row: (pd.to_datetime(f"{row[game_year_column]}-01-01") - row[birth_date_column]).days // 365, axis=1
        )
    
    return df

# Assuming your combined dataset is 'df'
df = calculate_age(df)

# Display the modified dataframe with the 'age_pitcher' and 'age_hitter' columns
print(df[['birthDate_pitcher', 'birthDate_batter', 'game_year', 'age_pitcher', 'age_batter']].head())
```

```python
import pandas as pd

def calculate_age_from_debut(df, game_year_column='game_year', debut_columns=['debut_pitcher', 'debut_batter']):
    """
    This function calculates the player's seasons played based on their debut date and the game_year.
    Adds new columns 'seasons_played_pitcher' and 'seasons_played_batter' to the DataFrame.
    """
    for debut_column in debut_columns:
        if debut_column in df.columns:
            # Convert debut column to datetime
            df[debut_column] = pd.to_datetime(df[debut_column], errors='coerce')

            # Create a new column for seasons played
            seasons_column = f"seasons_played_{debut_column.split('_')[-1]}"
            df[seasons_column] = df.apply(
                lambda row: (pd.to_datetime(f"{row[game_year_column]}-01-01") - row[debut_column]).days // 365
                if pd.notnull(row[debut_column]) else None, 
                axis=1
            )
    
    return df

# Apply the function to df
df = calculate_age_from_debut(df)

# Display the modified DataFrame with the new 'seasons_played' columns
print(df[['debut_pitcher', 'debut_batter', 'game_year', 'seasons_played_pitcher', 'seasons_played_batter']].head())
```

```python
def remove_date_columns(df):
    """
    Removes all columns from the DataFrame that have a datetime-like data type.
    """
    # Identify columns with datetime-like data types
    date_columns = df.select_dtypes(include=['datetime', 'datetime64', 'timedelta', 'object']).columns
    
    # Remove the date columns from the DataFrame
    df.drop(columns=date_columns, inplace=True)
    
    return df

# Apply the function to the 'df' DataFrame
df = remove_date_columns(df)

# Display the modified DataFrame
print(df.head())
```

## Data Splitting for Training the Model

```python
pa_target = df.pa
pa_features = df.drop(['bf', 'pa'], axis=1)
bf_target = df.bf
bf_features = df.drop(['bf', 'pa'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
# Perform train-test splits
pa_x_train, pa_x_test, pa_y_train, pa_y_test = train_test_split(
    pa_features, pa_target, train_size=0.7, test_size=0.3, 
    random_state=138496
)

bf_x_train, bf_x_test, bf_y_train, bf_y_test = train_test_split(
    bf_features, bf_target, train_size=0.7, test_size=0.3, 
    random_state=138496
)
```

## Model Creation

Here I use a cross validation score to figure out which regression model is best suited for the data and the desired outcome.

The best model for the data is Random Forest.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# 1. Define models
models = [
    ("Linear Regression", LinearRegression()),
    ("Ridge Regression", Ridge()),
    ("Lasso Regression", Lasso()),
    ("Random Forest", RandomForestRegressor(n_estimators=100, random_state=42)),
    ("Gradient Boosting", GradientBoostingRegressor(n_estimators=100, random_state=42)),
    ("AdaBoost", AdaBoostRegressor(n_estimators=100, random_state=42)),
]

# 2. Setup the Grid (Rows = Number of models, Cols = 2)
# We set a large height (5 inches per model) to ensure they are legible
fig, axes = plt.subplots(nrows=len(models), ncols=2, figsize=(15, 5 * len(models)))

best_model_name, best_model, best_rmse = None, None, float("inf")

# 3. Loop and Plot
for i, (name, model) in enumerate(models):
    # Train and Predict (Assuming pa_x_train, etc. are already defined in your memory)
    model.fit(pa_x_train, pa_y_train)
    y_pred = model.predict(pa_x_test)
    
    # Metrics
    rmse = np.sqrt(mean_squared_error(pa_y_test, y_pred))
    r2 = r2_score(pa_y_test, y_pred)
    if rmse < best_rmse:
        best_rmse, best_model_name, best_model = rmse, name, model

    # Column 1: Predicted vs Actual
    ax_scatter = axes[i, 0]
    sns.scatterplot(x=pa_y_test, y=y_pred, alpha=0.5, ax=ax_scatter)
    ax_scatter.plot([pa_y_test.min(), pa_y_test.max()], [pa_y_test.min(), pa_y_test.max()], 'r--')
    ax_scatter.set_title(f"{name}: Predicted vs Actual (RÂ²: {r2:.4f})")

    # Column 2: Residuals
    ax_resid = axes[i, 1]
    sns.histplot(pa_y_test - y_pred, kde=True, bins=30, ax=ax_resid)
    ax_resid.axvline(x=0, color='r', linestyle='--')
    ax_resid.set_title(f"{name}: Residuals (RMSE: {rmse:.4f})")

# 4. Final Formatting
#plt.tight_layout()
plt.savefig(r'C:\Users\cboat\OneDrive\Documents\availability_models_pa.png')
plt.show()

print(f"Best model: {best_model_name} with RMSE: {best_rmse:.4f}")
```

```python
# 1. Extract feature importances
# This works for Random Forest, Gradient Boosting, and AdaBoost
if hasattr(best_model, 'feature_importances_'):
    importances = best_model.feature_importances_
    feature_names = pa_x_train.columns
    
    # Create a DataFrame for easy sorting
    feature_importance_df = pd.DataFrame({
        'Feature': feature_names,
        'Importance': importances
    }).sort_values(by='Importance', ascending=False)

    # 2. Plotting
    plt.figure(figsize=(12, 8))
    sns.barplot(
        x='Importance', 
        y='Feature', 
        data=feature_importance_df.head(15), # Show top 15 drivers
        palette='viridis'
    )

    # Add labels and title
    plt.title(f'Top 15 Driving Factors of Hitters Playing Time ({best_model_name})', fontsize=16)
    plt.xlabel('Relative Importance Score', fontsize=12)
    plt.ylabel('Player Stats / Features', fontsize=12)
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    
    # Ensure no clipping of labels
    plt.tight_layout()
    plt.savefig(r'C:\Users\cboat\OneDrive\Documents\Feature_Importance_Hitters.png')
    plt.show()

elif hasattr(best_model, 'coef_'):
    # If the best model was Linear, Ridge, or Lasso
    importances = np.abs(best_model.coef_)
    feature_names = pa_x_train.columns
    # (Similar plotting logic as above for coefficients)
    print("Linear model detected: Plotting absolute coefficient weights.")
else:
    print("This model type does not support standard feature importance visualization.")
```
